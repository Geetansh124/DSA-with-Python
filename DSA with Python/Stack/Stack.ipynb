Let's Learn Stack

```python
class stack :
    def __init__(self):
        self.stack = []
    def length(self):
        return len(self.stack)
    def push(self, value):
        self.stack.insert(0, value)

    def isEmpty(self):
        if len(self.stack) == 0:
            raise Exception("Stack is empty")
        else:
            return self.stack[0]

    def peek(self):
        if len(self.stack) == 0:
            raise Exception("Stack is empty")
        else:
            return self.stack[0]

    def pop(self):
        if len(self.stack) == 0:
            raise Exception("Stack is empty")
        else:
            return self.stack.pop(0)

stk = stack()
stk.push(10)
stk.push(20)
stk.push(30)


print(stk.pop())
print(stk.pop())
print(stk.pop())


```
