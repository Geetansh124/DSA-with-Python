**Greedy Algorithm**

```python
def Max_min_diff(arr):
    arr.sort()
    n = len(arr)
    mid = n//2
    max = 0
    min = 0
    j = n - 1
    
    for i in range(mid):
        max = max +abs(arr[i] - arr[j])
        j = j - 1
        min = min + abs(arr[2*i] - arr[2*i+1])
    print("Max Difference:", max)    
    print("Min Difference:", min)    
    
arr = [12,5,25,10,2,15,8,30]
Max_min_diff(arr)    
```

```python
def minCoins(amount):
    coins = [500, 100, 50, 20, 10, 5, 2, 1]
    result = []

    for coin in coins:
        while amount >= coin:
            amount -= coin
            result.append(coin)

    return result

print(minCoins(1024))
```
